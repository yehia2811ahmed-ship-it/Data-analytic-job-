```python
import pandas as pd
import numpy as np

# استبدل المسار بالمسار الصحيح للملف على جهازك
df = pd.read_csv('Job Posts.csv')

# عرض أول 5 صفوف عشان نتأكد إن الملف اتمرر صح
df.head()
```

```python
# 1. Get the shape of the dataframe (rows, columns)
print("Data shape (rows, columns):", df.shape)
print("-" * 50)

# 2. Get data types and non-null counts
print("Dataframe information:")
df.info()
print("-" * 50)

# 3. Count missing values (Nulls) per column
print("Missing values per column:")
print(df.isnull().sum())
print("-" * 50)

# 4. Count total duplicate rows
print("Total duplicate rows:", df.duplicated().sum())
```

```python
# 1. Drop the unnecessary 'Unnamed: 0' column
df = df.drop(columns=['Unnamed: 0'])


# 3. Check the updated dataframe columns
df.head()
```

```python
# 1. Count exact technical skills demanded in descriptions
skills = ['sql', 'python', 'excel', 'tableau', 'power bi', 'sas', 'spark']
skill_counts = {}

for skill in skills:
    # Check if the skill exists as a word in the description
    skill_counts[skill] = df['Job_Description'].str.contains(rf'\b{skill}\b', case=False, regex=True).sum()

print("Exact Skills Demand:")
for skill, count in skill_counts.items():
    print(f"- {skill.upper()}: {count} jobs")

print("-" * 50)

# 2. Get the Top 10 most demanded Job Titles
print("Top 10 Most Demanded Job Titles:")
print(df['Job_Title'].value_counts().head(10))
```

```python
# 1. Total Job Posts
total_jobs = len(df)

# 2. Total Unique Job Titles
unique_titles = df['Job_Title'].nunique()

# 3. Top Skill and its percentage (SQL = 4055)
top_skill_name = "SQL"
top_skill_count = 4055
top_skill_pct = (top_skill_count / total_jobs) * 100

# 4. Top Job Role and its percentage (Data Scientist = 372)
top_role_name = "Data Scientist"
top_role_count = 372
top_role_pct = (top_role_count / total_jobs) * 100

# Print the KPI Dashboard Cards
print("=" * 40)
print("   DASHBOARD KPI CARDS DATA   ")
print("=" * 40)
print(f"CARD 1 -> Total Job Posts: {total_jobs:,}")
print(f"CARD 2 -> Unique Job Titles: {unique_titles:,}")
print(f"CARD 3 -> Top Demanded Skill: {top_skill_name} ({top_skill_pct:.1f}% of jobs)")
print(f"CARD 4 -> Top Job Role: {top_role_name} ({top_role_pct:.1f}% of market)")
print("=" * 40)
```

```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import seaborn as sns
import pandas as pd

# 1. Prepare data
skills_data = {'SQL': 4055, 'PYTHON': 3498, 'EXCEL': 1479, 'SPARK': 1406, 'TABLEAU': 1328, 'SAS': 627, 'POWER BI': 460}
skills_series = pd.Series(skills_data).sort_values(ascending=False)

# 2. Set up layout with GridSpec (2 rows: Row 0 for Cards, Row 1 for Chart)
fig = plt.figure(figsize=(15, 10), facecolor='#0f172a') # Dark modern background
gs = gridspec.GridSpec(2, 4, height_ratios=[1, 4]) # 4 columns for the 4 cards

# Define custom styling for cards
card_facecolor = '#1e293b'
text_color = '#ffffff'
accent_color = '#38bdf8'

# ----------------- ROW 0: THE 4 KPI CARDS -----------------
kpis = [
    {"title": "Total Job Posts", "value": "9,067"},
    {"title": "Unique Job Titles", "value": "6,385"},
    {"title": "Top Demand Skill", "value": "SQL (44.7%)"},
    {"title": "Top Job Role", "value": "Data Scientist (4.1%)"}
]

for i, kpi in enumerate(kpis):
    ax_card = fig.add_subplot(gs[0, i])
    ax_card.set_facecolor(card_facecolor)
    ax_card.set_xticks([])
    ax_card.set_yticks([])
    for spine in ax_card.spines.values():
        spine.set_color(accent_color)
        spine.set_linewidth(1.5)
        
    # Add Text to Cards
    ax_card.text(0.5, 0.65, kpi["title"], color='#94a3b8', fontsize=11, ha='center', va='center', fontweight='bold')
    ax_card.text(0.5, 0.3, kpi["value"], color=text_color, fontsize=14, ha='center', va='center', fontweight='bold')

# ----------------- ROW 1: THE CHART -----------------
# Spans across all 4 columns of the second row
ax_chart = fig.add_subplot(gs[1, :])
ax_chart.set_facecolor('#1e293b')

# Plotting Top Demanded Skills
sns.barplot(x=skills_series.values, y=skills_series.index, ax=ax_chart, palette="coolwarm")

# Customizing Chart Aesthetics
ax_chart.set_title('DEMAND FOR TECHNICAL SKILLS', fontsize=16, fontweight='bold', color=text_color, pad=15)
ax_chart.set_xlabel('Number of Job Posts', color='#94a3b8', fontsize=12, labelpad=10)
ax_chart.set_ylabel('Skills / Tools', color='#94a3b8', fontsize=12)
ax_chart.tick_params(colors=text_color, labelsize=11)
ax_chart.grid(axis='x', linestyle='--', alpha=0.3, color='#94a3b8')

for spine in ax_chart.spines.values():
    spine.set_color('#334155')

plt.tight_layout()
plt.show()
```

# 📊 Data Analytics Job Market - Executive Summary Report

## 1. Project Overview
This project aims to clean, explore, and analyze a dataset containing **9,067** job postings in the Data Analytics domain. The objective is to extract key marketplace insights, identify industry demands, and visualize high-priority metrics for data professionals.

---

## 2. Data Pipeline & Cleaning Methodology
* **Data Ingestion:** Loaded the raw `Job Posts.csv` containing 9,067 records and 4 initial attributes (`Unnamed: 0`, `Job_Title`, `Job_Description`, `Job_Link`).
* **Data Cleaning & Auditing:** 
  * Checked for missing entries; the dataset was verified to have **0 missing values (Nulls)** across all columns.
  * Checked for duplication; **0 identical duplicate rows** were found.
  * Structural adjustment: Dropped the redundant index column (`Unnamed: 0`) to optimize memory and processing efficiency.

---

## 3. Exploratory Data Analysis & Feature Extraction
* **Targeted Keyword Mining:** Since standard tokenization generated generic English stop words, a targeted programmatic regex search was applied directly to the `Job_Description` text column to accurately capture the exact frequencies of tech stack requirements.
* **Granularity Check:** Evaluated unique role variations, showing high fragmentation in naming conventions with thousands of distinct titles.

---

## 4. Key Performance Indicators (KPIs)
* **Total Marketplace Volume:** 9,067 active job postings.
* **Role Diversity:** 6,385 unique job title variations, showing a highly customized hiring market.
* **Dominant Technical Skill:** **SQL** is the most critical market requirement, appearing in **44.7%** of all job descriptions (4,055 posts).
* **Top Job Role Title:** **Data Scientist** stands out as the most standardized job title, holding **4.1%** of the total market share (372 posts).

---

## 5. Visual Dashboard Insights
* **The Tech Stack Spectrum:** A clear gap exists between core foundational skills and secondary tools. **SQL (4,055)** and **Python (3,498)** form the mandatory baseline for the data industry, significantly outperforming analytical tools like Excel, Spark, and Tableau, while Power BI holds the specialized tail end of the spectrum within this specific dataset.
